```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import loadmat
%matplotlib inline
```

```python
def find_closest_centroids(x,centroids):
    n = len(x)
    k = len(centroids)
    idx = np.zeros(n)
    for i in range(n):
        min_dist = np.sum((x[i,:] - centroids[0,:])**2)
        idx[i] = 0
        for j in range(1,k):
            dist = np.sum((x[i,:] - centroids[j,:])**2)
            if dist < min_dist:
                min_dist = dist
                idx[i] = j
    return idx
```

```python
data = loadmat("ex7data2.mat")
```

```python
x_train = data["X"]
```

```python
centroids = np.array([[3,3],[6,2],[8,5]])
```

```python
idx = find_closest_centroids(x_train,centroids)
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(x_train[:,0],x_train[:,1],marker="o")
ax.scatter(centroids[:,0],centroids[:,1], marker="x")
```

```python
def compute_centroids(x,idx,k):
    n,m = x.shape
    centroids = np.zeros((k,m))
    for i in range(k):
        centroids[i,:] = np.mean(x[idx==i],axis=0)
    return centroids
```

```python
idx = find_closest_centroids(x_train,centroids)
centroids = compute_centroids(x_train,idx,3)
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(x_train[:,0],x_train[:,1],marker="o")
ax.scatter(centroids[:,0],centroids[:,1], marker="x")
```

```python
def run_k_means(x,centroids, max_iters):
    n,m = x.shape
    k = centroids.shape[0]
    centroids_dic = {}
    for j in range(k):
        centroids_dic[str(j)] = [list(centroids[j,:])]
    idx = np.zeros(n)
    for i in range(max_iters):
        idx = find_closest_centroids(x,centroids)
        centroids = compute_centroids(x,idx,k)
        for j in range(k):
            centroids_dic[str(j)].append(list(centroids[j,:]))
    return idx,centroids,centroids_dic
```

```python
def init_centroids(x,k):
    n,m = x.shape
    centroids = np.zeros((k,m))
    idx = np.random.randint(0,n,k)
    
    for i in range(k):
        centroids[i,:] = x[idx[i]]
        
    return centroids
```

```python
centroids = init_centroids(x_train,3)
ids, centroids,centroids_dic = run_k_means(x_train, centroids, 10)
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(x_train[idx==0,0],x_train[idx==0,1],marker="o",color="r")
ax.scatter(x_train[idx==1,0],x_train[idx==1,1],marker="o",color="b")
ax.scatter(x_train[idx==2,0],x_train[idx==2,1],marker="o",color="k")
for k in range(len(centroids)):
    centroid = np.array(centroids_dic[str(k)])
    ax.plot(centroid[:,0],centroid[:,1],"gx-")
```
